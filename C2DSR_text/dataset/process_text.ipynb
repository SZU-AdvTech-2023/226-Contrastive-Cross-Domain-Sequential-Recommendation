```python
# 处理文本
# 文本信息需要哪些？
# title brand category
# 加载mapping 文件
mapping = pd.read_csv("processed_data_all/mapping_items.csv",header=None,names=["new","origin"],sep=",")
mapping.head()
```

```python
import pandas as pd
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

movie_df = getDF('amazon/meta_Movies_and_TV.json.gz')
cd_df = getDF('amazon/meta_CDs_and_Vinyl.json.gz')
book_df = getDF('amazon/meta_Books.json.gz')
```

```python
print(book_df[book_df["asin"] == "B00M111QCM"].categories)
```

```python
metadata = pd.concat([movie_df,cd_df,book_df],axis=0)

```

```python
metadata
```

```python
metadata.drop_duplicates(subset=["asin"],keep="first",inplace=True)
```

```python
merge_data = pd.merge(left=mapping,right=metadata,how="left",left_on="origin",right_on="asin")
```

```python
merge_data.head()
```

```python
# 选出title brand categories
select_data = merge_data[["new","origin","categories","title","brand"]]
select_data
```

```python
# 对以上信息进行清洗，将Nan数据置为空”“
select_data.fillna(" ",inplace=True)
import html
import re

def clean_text(raw_text):
    if isinstance(raw_text, list):
        cleaned_text = ' '.join("%s" % item for item in raw_text)
    elif isinstance(raw_text, dict):
        cleaned_text = str(raw_text)
    else:
        cleaned_text = raw_text
    # 解释 HTML 实体引用
    #作用是：把类似&#39;这样的html实体转换成'，以此类推
    cleaned_text = html.unescape(cleaned_text)
    # 解释：去除一些特殊符号比如\u3000
    cleaned_text = re.sub(r'["\n\r]*', '', cleaned_text)
    index = -1
    # 解释while作用是：去除句子开头的.号
    # 例如：.hello world -> hello world
    while -index < len(cleaned_text) and cleaned_text[index] == '.':
        index -= 1
    index += 1
    # 如果是空字符串，那么就返回空字符串
    if index == 0:
        cleaned_text = cleaned_text + '.'
    else:
        # 如果不是空字符串，那么就去除开头的.号
        cleaned_text = cleaned_text[:index] + '.'
    if len(cleaned_text) >= 2000:
        cleaned_text = ''
    return cleaned_text

# 定义df的apply函数,对categories title brand进行清洗，并将三者拼接起来
def clean_df(df):
    df['categories'] = df['categories'].apply(clean_text)
    df['title'] = df['title'].apply(clean_text)
    df['brand'] = df['brand'].apply(clean_text)
    df['text'] = df['title'] + ' ' + df['brand'] + ' ' + df['categories']
    return df

# 对select_data进行清洗
clean_data = clean_df(select_data)
clean_data.head()
```

```python
output_data = clean_data[['new','origin','text']]
# 输出到文件中 映射ID， 原始ID，文本信息
output_data.to_csv("processed_data_all/mapping_items4text.csv",index=False,sep="\t",header=None)
 
```
